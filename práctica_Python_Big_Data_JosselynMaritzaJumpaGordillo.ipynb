```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra

# B. Una función que calcule el precio total de la compra con un descuento del 10%

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

# D. Una función que calcule el precio total aplicando el IVA (21%)

# E.lista los artículos cuyo precio es superior a 20€
```

```python
# A. Una función que calcule el precio total de la compra (imperativa)

def total_compra(compra, articulos):
    result = 0
    for item in compra:
        if item in articulos:
            result += articulos[item]
    return result

precio_compra = total_compra(compra,articulos)

precio_compra
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10% (funcional)

discount_10 = map(lambda precio: precio - (precio*0.10), [precio_compra])

print(list(discount_10))
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10% (imperativa)


def total_discount(total):
    percentage = 10
    discount = total *(percentage /100)
    new_total = total - discount
    return new_total
    
precio_total = total_discount(precio_compra)

precio_total
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€ (funcional)

#ejemplo 
discount_over_100 = map(lambda precio: precio - (precio*0.10) if precio > 100 else precio, [precio_compra])

print(list(discount_over_100))
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€ (imperativa)

def si_compra_100(precio_compra):
    if precio_compra > 100:
        return total_discount(precio_compra)
       
    else:
        return precio_compra
    

si_compra_100(precio_compra)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%) (funcional)

iva_incluido = map(lambda precio: precio + (precio*0.21), [precio_compra])
    
print(list(iva_incluido))    
```

```python
# E.lista los artículos cuyo precio es superior a 20€  (imperativa)

def items_over_20(articulos):
    items_20 = []
    for item in articulos:
        if articulos[item] > 20:
            items_20.append(item)
               
    return items_20
            



items_over_20(articulos)
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]
```

```python
def alumnos_promedios(curso):
    promedio_calificaciones = lambda calificaciones: reduce(lambda x, y: x + y, calificaciones) / len(calificaciones)
    lista_alumnos = []

    for alumno in alumnos:
        alumno_list = list(alumno)
        if alumno_list[2]== str(curso):
            promedio = promedio_calificaciones(alumno_list[3])
            info_alumn = [alumno_list[0:2], promedio]
            lista_alumnos.append(info_alumn)
    return(lista_alumnos)
            
 
           
#ejemplo

alumnos_promedios(1)
```
